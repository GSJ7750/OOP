```python
########## 4 - 1 ###########
def column(a_list, num):
    if num == 0:###number of row가 바뀔 위험을 배제
        print(a_list[0][1], "~", a_list[0][-1], "사이의 숫자를 입력 해 주세요.")
        return
    if num <= -1:####column의 범위보다 underflow인 숫자를 입력 시 일반 list와 똑같이 기능하도록 함
        num = a_list[0][num]
        
    attList = []
    print("Attribute of column",num)
    for i in range(len(a_list)-1):#헤더 부분을 출력하지 않기 위해 -1
        attList.append(a_list[i+1][num])
        #print(a_list[i+1][0], attList[i])
        
    return attList
print(column(a_list, -3))
print(column(a_list, 2))

temp_a_list = copy.deepcopy(a_list)#a_list는 실제로 변하지 않아야 하기 때문에 임시로 깊은 복사를 해 둠.

########## 4 - 2  ###########
def columnReverse(a_list, num): 
    if num == 0:#number of row가 바뀌지 않을 것
        print(a_list[0][1], "~", a_list[0][-1], "사이의 숫자를 입력 해 주세요.")
        return
    if num <= -1:#column의 범위보다 underflow인 숫자를 입력 시 일반 list와 똑같이 기능하도록 함
        num = a_list[0][num]
        
    attList = []
    tempList = []
    for i in range(len(a_list)-1):#헤더 부분을 출력하지 않기 위해 -1
        attList.append(a_list[i+1][num])
    attList.reverse()
    
    for i in range(len(attList)):
        tempList.append(a_list[i+1][num+1:])#헤더와 바뀌는 열은 제외하고 임시로 저장
        del(a_list[i+1][num:])#바뀌는 열 부터 마지막 열 까지 삭제
    
    for i in range(len(attList)):
        a_list[i+1].append(attList[i])
        a_list[i+1].extend(tempList[i])#list 형태를 append 하게 되면 list형태로 추가되기 때문에 내부 요소만 추가.
    
    return a_list

########## 4 - 3 ###########
c_list = columnReverse(a_list, -3)
a_list = temp_a_list
```

```python
########## 5 - 1 ###########
def divide_train_test(in_list, prop):
    num = len(in_list)#리스트의 갯수를 받아옴
    trainNum = int(num*prop)+1#header 제외
    train_data = in_list[1:trainNum]#header+1 : trainNum
    test_data = in_list[trainNum:]#trainNum : -1
    
    return train_data, test_data


########## 5 - 2 ###########
train_data, test_data = divide_train_test(in_list=c_list, prop=0.6)
print(len(train_data))##prop만큼의 비율로 분할 되었는지 확인
print(len(test_data),'\n')
print(train_data)##실제 데이터
print(test_data, '\n\n')

train_data, test_data = divide_train_test(a_list, 0.8)
print(len(train_data))##
print(len(test_data), '\n')
print(train_data)
print(test_data)
```

```python
########## 6 - 1 ###########
def min_max_avg(in_list):
    return [min(in_list), max(in_list), sum(in_list)/len(in_list)]

########## 6 - 2 ###########
randList = []
for i in range(20):
    randList.append(int(random.randrange(-50, 50)))  
    
MMA = min_max_avg(randList)
MM = MMA[:2]
print(min_max_avg([1,6,2,8,3,5,-4,2]))
print(MMA,'\n\n')

########## 6 - 3 ###########
def equ_discretize(_range, intervals):
    output = []
    _min, _max = _range[0], _range[1]
    a_min, a_max=  abs(_min), abs(_max)
    distance = a_min + a_max#절댓값으로 변환하여 음수, 양수 관계 없이 거리를 구함
    i_distance = int(distance/intervals)# 1회 당 더해질 숫자
    for i in range(intervals):
        output.append([_min, _min+i_distance])
        _min += i_distance
    return output

########## 6 - 4 ###########
print(equ_discretize([-4,8],3))
print(equ_discretize([-5,10],5))
print(equ_discretize([-20,100],4))
print(equ_discretize([10,50],2))
```

```python
def E(a, n):
    _list = []
    abssum = abs(a[0])+abs(a[1])
    dist = abssum/n
    for i in range(n):
        _list.append(dist*i)
    
    return _list
```

```python
def divide(inlist, prop):
    train = int(len(inlist)*prop)
    train_data = inlist[:train][:]
    test_data = inlist[train:][:]
    return train_data, test_data
```
